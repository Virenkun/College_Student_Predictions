Preprocessing the Data Our goal with prerocessing is to change our numerical fields that have a value like GradeID to a numerical only value in a way that we preserve that distance in a meningful way. Additionally, we want to assign our three classes to numerical outcomes with a preserved distance. There are a couple of ways to do this. We went with setting L = -1, M = 0, and H = 1. Additionally, you could set each to the middle value of their category on the 100% scale (L = 34.5, M = 79.5, and H = 95). We chose to preserve the distance between the categorical values. Additionally, we decided to scale our numerical fields so that they would be more meaningful when compared together. For this we used scikit learnâ€™s built in pre-processing scaling ability.

```python
X = df.drop(['Class'], axis=1)
y = df['Class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
trf1 = ColumnTransformer(transformers=[('tnf1',OneHotEncoder(sparse=False,drop='first'), [0,1,2,3,5,6,7,8,13,14,15]),
                                      ('tnf2', OrdinalEncoder(categories=[["G-01", "G-02","G-03" , "G-04", "G-05" , "G-06" ,"G-07" ,"G-08" ,"G-09" ,"G-10" ,"G-11" ,"G-12"]]), [4]), ('Scaler', StandardScaler(), [9,10,11,12])], remainder='passthrough')
```

```python
X_train_trf = trf1.fit_transform(X_train)
```

```python
X_test_trf = trf1.transform(X_test)
```

```python
lb = LabelEncoder()
```

```python
y_train_trf = lb.fit_transform(y_train)
```

```python
y_test_trf = lb.transform(y_test)
```

```python
y_train_trf
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn import metrics
```

```python
# Define the models to compare
models = [
    LogisticRegression(max_iter=300),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    SVC(),
    GradientBoostingClassifier(),
    MLPClassifier(max_iter=2000)
]
```

```python
metric = 'accuracy'

for model in models:
    scores = cross_val_score(model, X_train_trf, y_train_trf, cv=5, scoring=metric)
    print(f"Model: {model.__class__.__name__}")
    print(f"Cross-validated {metric} scores: {scores}")
    print(f"Mean {metric} score: {scores.mean():.4f}")
    print("---")

    # Fit the model on the entire dataset
    model.fit(X_train_trf, y_train_trf)
    # Use the fitted model for predictions
    y_pred = model.predict(X_test_trf)
    # Calculate evaluation metrics on the entire dataset
    accuracy = metrics.accuracy_score(y_test_trf, y_pred)
    precision = metrics.precision_score(y_test_trf, y_pred, average='weighted')
    recall = metrics.recall_score(y_test_trf, y_pred, average='weighted')
    f1 = metrics.f1_score(y_test_trf, y_pred, average='weighted')
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("---\n")
```
